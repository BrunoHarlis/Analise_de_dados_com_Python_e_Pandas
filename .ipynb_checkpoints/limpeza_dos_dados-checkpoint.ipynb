## LIMPEZA DOS DADOS

```python
# Importando Pandas
import pandas as pd
import pandera as pa # biblioteca para validacao de documentos
```

```python
# Criando um DataFrame de um arquivo .csv
df = pd.read_csv('marketing_campaign.csv', sep='\t')
df.head()
```

```python
df.tail()
```

```python
# Verifica se a coluna ID possui valores unicos
df.ID.is_unique
```

```python
# Colocando a coluna ID como indice
df.set_index('ID', inplace=True)
df.head()
```

```python
# Retorna ao indice padrao
#df.reset_index(drop=True, inplace=True)
#df.head()
```

```python
# Mostra quantos valores nulos possuem em cada coluna
df.isna().sum()
```

```python
# Exclue as linhas que possuem valores nulos
df.dropna(inplace=True)
```

## TRANSFORMACAO DOS DADOS

```python
# Renomeando colunas 
df = df.rename(columns={
    'Year_Birth':'Ano_Nascimento', 
    'Education':'Formacao',
    'Marital_Status':'Estado_Civil',
    'Income':'Renda',
    'Kidhome':'Crianca',
    'Teenhome':'Adolecente',
    'Dt_Customer':'Dt_Cliente',
    'Recency':'Dias_Ultima_Compra',
    'MntWines':'QtdVinhos',
    'MntFruits':'QtdFrutas',
    'MntMeatProducts':'QtdCarnes',
    'MntFishProducts':'QtdPeixes',
    'MntSweetProducts':'QtdDoces',
    'MntGoldProds':'Qtd_prod_Ouro',
    'NumDealsPurchases':'Compra_Oferta',
    'NumWebPurchases':'Compra_Web',
    'NumCatalogPurchases':'Compra_Catalogo',
    'NumStorePurchases':'Compra_loja',
    'NumWebVisitsMonth':'Visitas_web_mes',
    'AcceptedCmp3':'Compra_Oferta_3',
    'AcceptedCmp4':'Compra_Oferta_4',
    'AcceptedCmp5':'Compra_Oferta_5',
    'AcceptedCmp1':'Compra_Oferta_1',
    'AcceptedCmp2':'Compra_Oferta_2',
    'Complain':'Reclamacoes',
    'Z_CostContact':'Custo_Contato_Cliente',
    'Z_Revenue':'Receita_Apos_Oferta',
    'Response':'Compra_Ultima_Oferta'})
df.head()
```

```python
# Criando uma coluna Idade
df['Idade'] = 2021 - df['Ano_Nascimento']
df.head()
```

```python
# mostrando a idade maxima
df.sort_values('Idade', ascending=False).head()
```

```python
df.loc[11004]
```

```python
# Removendo clientes com idade acima de 100 anos
for x in df.index:
    if df.loc[x, 'Idade'] > 100:
        df.drop(x, inplace = True)
          
df.sort_values('Idade', ascending=False).head()
```

```python
# Mostrando os tipos de dado das colunas
df.dtypes
```

```python
# Mudando o tipo de dado da coluna Dt_Cliente de object para datatime64
df['Dt_Cliente'] = pd.to_datetime(df['Dt_Cliente'], errors ='coerce')
df.dtypes
```

## VALIDACAO DOS DADOS

```python
schema = pa.DataFrameSchema(
    columns = {
        'Ano_Nascimento':pa.Column(pa.Int),
        'Formacao':pa.Column(pa.String),
        'Estado_Civil':pa.Column(pa.String),
        'Renda':pa.Column(pa.Float),
        'Crianca':pa.Column(pa.Int),
        'Adolecente':pa.Column(pa.Int),
        'Dt_Cliente':pa.Column(pa.DateTime),
        'Dias_Ultima_Compra':pa.Column(pa.Int),
        'QtdVinhos':pa.Column(pa.Int),
        'QtdFrutas':pa.Column(pa.Int),
        'QtdCarnes':pa.Column(pa.Int),
        'QtdPeixes':pa.Column(pa.Int),
        'QtdDoces':pa.Column(pa.Int),
        'Qtd_prod_Ouro':pa.Column(pa.Int),
        'Compra_Oferta':pa.Column(pa.Int),
        'Compra_Web':pa.Column(pa.Int),
        'Compra_Catalogo':pa.Column(pa.Int),
        'Compra_loja':pa.Column(pa.Int),
        'Visitas_web_mes':pa.Column(pa.Int),
        'Compra_Oferta_3':pa.Column(pa.Int),
        'Compra_Oferta_4':pa.Column(pa.Int),
        'Compra_Oferta_5':pa.Column(pa.Int),
        'Compra_Oferta_1':pa.Column(pa.Int),
        'Compra_Oferta_2':pa.Column(pa.Int),
        'Reclamacoes':pa.Column(pa.Int),
        'Custo_Contato_Cliente':pa.Column(pa.Int),
        'Receita_Apos_Oferta':pa.Column(pa.Int),
        'Compra_Ultima_Oferta':pa.Column(pa.Int),
    }
)
```

```python
# Validando os dados
schema.validate(df)
```

## ANALIZE DOS DADOS

```python
# Quais tipos de clientes gastam mais com vinho?
import matplotlib.pyplot as plt
df.groupby(df['Estado_Civil'])['QtdVinhos'].sum().plot.bar();
```

```python
# Quais tipos de clientes gastam mais com frutas?
df.groupby(df['Estado_Civil'])['QtdFrutas'].sum().plot.bar();
```

```python
# Quais clientes compram mais pela web?
df.groupby(df['Estado_Civil'])['Compra_Web'].sum().plot.bar();
```

```python
# Quantos clientes temos de cada Estado Civil?
df.groupby(['Estado_Civil']).size()
```

```python
# Criando a coluna Total_Gasto
df['Total_Gasto'] = df['QtdVinhos'] + df['QtdFrutas'] + df['QtdCarnes'] + df['QtdPeixes'] + df['QtdDoces']
df.sample(5)
```

```python
# Qual a idade dos clientes que preferem comprar na loja Fisica?
df.groupby(['Idade'])['Compra_loja'].sum().plot.barh();
```

```python
# Salvando a imagem em formato png
df.groupby(['Idade'])['Compra_loja'].sum().plot.barh()
plt.xlabel('Valor em compras')
plt.ylabel('Ano Nascimento')
plt.savefig('Idade Cliente que mais compram')
```
